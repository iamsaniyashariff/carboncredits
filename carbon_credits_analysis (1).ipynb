```python
missing_data = df.isnull().sum()

missing_percentage = (missing_data / len(df)) * 100

print("Missing data counts:\n", missing_data)
print("Missing data percentage:\n", missing_percentage)
```

```python
# dropping columns with > 90% missing data
high_missing_threshold = 90
columns_to_drop = missing_percentage[missing_percentage > high_missing_threshold].index
df_cleaned = df.drop(columns=columns_to_drop)

# imputing missing values for numerical columns (< 30% missing)
numerical_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_cols:
    if missing_percentage[col] < 30:
        df_cleaned[col].fillna(df_cleaned[col].median(), inplace=True)

# imputing missing values for categorical columns (< 30% missing)
categorical_cols = df_cleaned.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if missing_percentage[col] < 30:
        df_cleaned[col].fillna(df_cleaned[col].mode()[0], inplace=True)

print("Remaining missing data counts:\n", df_cleaned.isnull().sum())
```

```python
label_encoder = LabelEncoder()
df['Voluntary Status'] = label_encoder.fit_transform(df['Voluntary Status'])

# cleaning the column names
df.columns = df.columns.str.strip()

# encoding
required_columns = ['Region', 'Country', 'State']

# checking
for col in required_columns:
    if col in df.columns:
        df = pd.get_dummies(df, columns=[col], drop_first=True)

# binary encoding
binary_encoder = BinaryEncoder(cols=['Project Developer'])
df = binary_encoder.fit_transform(df)
```

```python
# frequency encoding
freq_encoding = df['Verifier'].value_counts().to_dict()
df['Verifier_Encoded'] = df['Verifier'].map(freq_encoding)

# target encoding
df['Type_Target_Encoded'] = df.groupby('Type')['Total Credits \r\nIssued'].transform('mean')
```

```python
# filtering columns for yearly credit-related data (credits issued, retired, remaining, and buffers by year)
yearly_columns = df.filter(regex=r'^\d{4}(\.\d+)?$').columns  # Matches columns with years (e.g., '1996', '1996.1', etc.)
yearly_credits = df[yearly_columns]  # Subset the yearly columns

# melting the data for yearly trends
trend_data = yearly_credits.melt(var_name='Year_Credit_Type', value_name='Credit Value')

# extracting 'Year' and 'Credit Type' from the melted data
trend_data['Year'] = trend_data['Year_Credit_Type'].str.extract(r'^(\d{4})').astype(int)  # Extract the year
trend_data['Credit Type'] = trend_data['Year_Credit_Type'].str.extract(r'(\.\d+)$')  # Extract credit type identifier (e.g., '.1', '.2', etc.)
trend_data['Credit Type'] = trend_data['Credit Type'].map({
    '.0': 'Credits Issued',
    '.0.1': 'Credits Retired',
    '.0.2': 'Credits Remaining',
    '.0.3': 'Buffer Pool Deposits',
}).fillna('Unknown')  # Map identifiers to human-readable types

# summary
yearly_trends = trend_data.groupby(['Year', 'Credit Type'])['Credit Value'].sum().reset_index()

# checking metadata
if 'Year' in df.columns:
    df['Year'] = pd.to_datetime(df['Year'], format='%Y', errors='coerce')


print(yearly_trends.head())
```

```python
df.columns = df.columns.str.replace(r'\s+', ' ', regex=True).str.strip()

print("Cleaned Column Names:")
print(df.columns)

# creating a 'Year' column by extracting the year from the year-based columns

year_columns = df.filter(regex=r'^\d{4}(\.\d+)?$').columns

# extraction
df['Year'] = pd.to_datetime(df[year_columns[0]].astype(str), format='%Y', errors='coerce')

# formatting
df['Year'] = pd.to_datetime(df['Year'], format='%Y', errors='coerce')

# ratio of Total Credits Retired to Total Credits Issued
df['Retired_to_Issued_Ratio'] = df['Total Credits Retired'] / df['Total Credits Issued']

# average annual emission reductions across available years
# by Year and calculate average emission reductions
average_annual_reduction = (
    df.groupby(df['Year'].dt.year)['Estimated Annual Emission Reductions']
    .mean()
    .reset_index()
    .rename(columns={'Year': 'Year', 'Estimated Annual Emission Reductions': 'Avg Annual Reduction'})
)

# merging
df = pd.merge(df, average_annual_reduction, on='Year', how='left')

# 3. aggregation of annual data into multi-year periods or significant policy milestones

df['Period'] = (df['Year'].dt.year // 5) * 5  # Grouping into 5-year periods

multi_year_aggregation = (
    df.groupby('Period')
    .agg({
        'Total Credits Issued': 'sum',
        'Total Credits Retired': 'sum',
        'Estimated Annual Emission Reductions': 'sum'
    })
    .reset_index()
    .rename(columns={
        'Total Credits Issued': 'Total Issued (5-Year)',
        'Total Credits Retired': 'Total Retired (5-Year)',
        'Estimated Annual Emission Reductions': 'Total Reductions (5-Year)'
    })
)

multi_year_aggregation.to_csv('multi_year_aggregation.csv', index=False)

print("Ratio of Retired to Issued Credits:")
print(df[['Year', 'Retired_to_Issued_Ratio']].head())

print("\nAverage Annual Emission Reductions:")
print(average_annual_reduction.head())

print("\nMulti-Year Aggregated Data:")
print(multi_year_aggregation.head())
```

```python
df.columns = df.columns.str.replace(r'\s+', ' ', regex=True).str.strip()

# ratio of Reversals Covered by Buffer Pool to Total Buffer Pool Deposits
df['Buffer_Pool_Efficiency_Ratio'] = df['Reversals Covered by Buffer Pool'] / df['Total Buffer Pool Deposits']

# projects with high reversal risks
#  high-risk projects ie those where the uncovered reversals exceed a threshold
threshold_ratio = 0.1
df['High_Reversal_Risk'] = (
    df['Reversals Not Covered by Buffer'] > threshold_ratio * df['Total Buffer Pool Deposits']
)

# summarize buffer pool efficiency and reversal risks across all projects
buffer_reversal_summary = df.agg({
    'Total Buffer Pool Deposits': 'sum',
    'Reversals Covered by Buffer Pool': 'sum',
    'Reversals Not Covered by Buffer': 'sum'
}).to_frame().T

# no division by zero in the summary ratio calculation
buffer_reversal_summary['Buffer_Pool_Efficiency_Ratio'] = (
    buffer_reversal_summary['Reversals Covered by Buffer Pool'] /
    buffer_reversal_summary['Total Buffer Pool Deposits']
)

# handling cases where 'Total Buffer Pool Deposits' could be zero
buffer_reversal_summary['Buffer_Pool_Efficiency_Ratio'] = buffer_reversal_summary['Buffer_Pool_Efficiency_Ratio'].replace([float('inf'), float('nan')], 0)

# high-risk projects for further review
high_risk_projects = df[df['High_Reversal_Risk']]
high_risk_projects.to_csv('high_risk_projects.csv', index=False)

print("Buffer Pool Efficiency Ratio (All Projects):")
print(buffer_reversal_summary)

print("\nProjects Identified with High Reversal Risks:")
print(high_risk_projects[['Project ID', 'Project Name', 'High_Reversal_Risk']].head())
```

```python
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import MultiLabelBinarizer

# sample column names
description_col = "Project Description"
notes_col = "Notes from Registry"

# combined text data from Project Description and Notes
df['Combined_Text'] = (
    df[description_col].fillna('') + ' ' + df[notes_col].fillna('')
)

# preprocessing
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)  # Remove special characters
    return text.strip()

df['Processed_Text'] = df['Combined_Text'].apply(preprocess_text)

# classify projects based on keywords
keywords = {
    'reforestation': ['reforestation', 'afforestation', 'forest', 'tree planting'],
    'methane': ['methane', 'landfill gas', 'CH4', 'biogas'],
    'renewable energy': ['solar', 'wind', 'renewable energy', 'hydropower'],
    'energy efficiency': ['efficiency', 'energy savings', 'retrofit'],
    'carbon sequestration': ['sequestration', 'carbon storage', 'soil carbon'],
}

# function for the above
def classify_text(text):
    categories = []
    for category, words in keywords.items():
        if any(word in text for word in words):
            categories.append(category)
    return categories

df['Categories'] = df['Processed_Text'].apply(classify_text)

# encoding
mlb = MultiLabelBinarizer()
categories_encoded = pd.DataFrame(mlb.fit_transform(df['Categories']),
                                   columns=mlb.classes_,
                                   index=df.index)

df = pd.concat([df, categories_encoded], axis=1)

# insights from text using word frequency
vectorizer = CountVectorizer(stop_words='english', max_features=20)  # top 20 common words
text_matrix = vectorizer.fit_transform(df['Processed_Text'])

word_freq_df = pd.DataFrame(text_matrix.toarray(), columns=vectorizer.get_feature_names_out())

word_summary = word_freq_df.sum().sort_values(ascending=False)

df.to_csv('processed_project_data.csv', index=False)
word_summary.to_csv('word_frequency_summary.csv', index=True)

print("Top Word Frequencies:")
print(word_summary.head(10))

print("\nSample of Classified Projects:")
print(df[['Project ID', 'Project Name', 'Categories']].head())
```

```python
import numpy as np

# columns for imputation based on your dataset
categorical_cols = [
    'Methodology Version', 'Region', 'Country', 'State', 'Voluntary Registry',
    'ARB / WA Project', 'Voluntary Status', 'Scope', 'Type', 'Reduction / Removal'
]

numerical_cols = [
    'Total Credits Issued', 'Total Credits Retired', 'Total Credits Remaining',
    'Total Buffer Pool Deposits', 'Reversals Covered by Buffer Pool',
    'Reversals Not Covered by Buffer', 'Estimated Annual Emission Reductions'
]

# annual data columns (time-series)
time_series_cols = [col for col in df.columns if col.startswith(('199', '200', '20')) and '.' in col]

missing_summary = df.isnull().sum().reset_index()
missing_summary.columns = ['Column', 'Missing Count']
missing_summary['Missing Percentage'] = (missing_summary['Missing Count'] / len(df)) * 100
print("Summary of Missing Values:")
print(missing_summary[missing_summary['Missing Count'] > 0].sort_values(by='Missing Percentage', ascending=False))

for col in categorical_cols:
    if col in df.columns:
        mode_value = df[col].mode()[0]  # Most frequent value
        df[col].fillna(mode_value, inplace=True)

# mean imputation for overall numerical data
for col in numerical_cols:
    if col in df.columns:
        mean_value = df[col].mean()
        df[col].fillna(mean_value, inplace=True)

# linear interpolation for time-series numerical data
for col in time_series_cols:
    if col in df.columns and df[col].dtype in [np.float64, np.int64]:
        df[col].interpolate(method='linear', inplace=True)

post_imputation_missing_summary = df.isnull().sum().reset_index()
post_imputation_missing_summary.columns = ['Column', 'Missing Count']
print("Post-Imputation Missing Values:")
print(post_imputation_missing_summary[post_imputation_missing_summary['Missing Count'] > 0].sort_values(by='Missing Count', ascending=False))



clustering_cols = [
    'Scope', 'Region', 'Country', 'State', 'Total Credits Issued',
    'Total Credits Retired', 'Total Credits Remaining', 'Estimated Annual Emission Reductions'
]

clustering_cols = [col for col in clustering_cols if col in df.columns]
```

```python
label_encoders = {}
for col in ['Scope', 'Region', 'Country', 'State']:
    if col in df.columns:
        le = LabelEncoder()
        df[col] = df[col].fillna('Unknown')  # Fill missing with 'Unknown' before encoding
        df[col] = le.fit_transform(df[col])
        label_encoders[col] = le

for col in ['Total Credits Issued', 'Total Credits Retired', 'Total Credits Remaining', 'Estimated Annual Emission Reductions']:
    if col in df.columns:
        df[col].fillna(df[col].mean(), inplace=True)

numerical_cols = [col for col in clustering_cols if col not in ['Scope', 'Region', 'Country', 'State']]
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

clustering_data = df[clustering_cols].copy()

inertia = []
K = range(2, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(clustering_data)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()

optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(clustering_data)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(clustering_data)
df['PCA1'] = pca_result[:, 0]
df['PCA2'] = pca_result[:, 1]

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='PCA1', y='PCA2', hue='Cluster', data=df, palette='Set2', s=100, alpha=0.8
)
plt.title('Clustering Results Visualized with PCA')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Cluster')
plt.show()

cluster_summary = df.groupby('Cluster')[clustering_cols].mean()
print("Cluster Characteristics Summary:")
print(cluster_summary)


```
